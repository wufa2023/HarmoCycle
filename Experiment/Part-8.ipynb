# ccRemover-mESC

```python
import os
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# ==============================
# Set working directory
# ==============================
os.chdir("/root/Cycle/Github")
print("Current working directory:", os.getcwd())

# ==============================
# Load processed CSV data
# ==============================
df = pd.read_csv('./Dataset/ccRemover_output/after_log_4datasets_final.csv', index_col=0)

# Convert DataFrame to AnnData object (transpose so cells are rows)
ref6 = sc.AnnData(df.T)

# Load reference dataset for cell metadata
leng_data = sc.read_h5ad('./Dataset/mESC.h5ad')

# Align cells with reference dataset
ref6 = ref6[leng_data.obs_names, ]
ref6.obs = leng_data.obs.copy()

# Copy to working AnnData object
adata = ref6.copy()

# ==============================
# Preprocessing
# ==============================
# Standard scaling (Z-score normalization)
sc.pp.scale(adata, max_value=None)  # max_value=None means no clipping

# ==============================
# Dimensionality reduction
# ==============================
sc.tl.pca(adata, n_comps=3)        # PCA with 3 components
sc.pp.neighbors(adata, n_pcs=3)    # Compute neighborhood graph using first 3 PCs
sc.tl.umap(adata)                  # UMAP embedding

# ==============================
# Compute silhouette score (based on PCA)
# ==============================
pca_data = adata.obsm['X_pca'][:, :3]
labels = adata.obs['stage']
silhouette_avg = silhouette_score(pca_data, labels)
print(f"Silhouette Score: {silhouette_avg}")

# ==============================
# Define custom color palette
# ==============================
custom_palette = {
    0: '#9FAFCB',  # Stage 0 color
    1: '#E19A90',  # Stage 1 color
    2: '#B9B1C5'   # Stage 2 color
}

# ==============================
# Plot UMAP
# ==============================
plt.figure(figsize=(10, 8))
sc.pl.umap(adata, color='stage', palette=custom_palette, show=False)
plt.title(f'UMAP projection - Silhouette Score: {silhouette_avg:.3f}')
plt.tight_layout()

# ==============================
# Ensure output folder exists
# ==============================
output_dir = './FigurePlot/CellCycleEffect'
os.makedirs(output_dir, exist_ok=True)

# Save figure as PDF
output_path = os.path.join(output_dir, 'CC_Effect_Remove_mESC_PC3.pdf')
plt.savefig(output_path, format='pdf', bbox_inches='tight')
print(f"Figure saved as PDF at: {output_path}")

# Uncomment to display interactively
# plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set font to commonly used academic style (DejaVu Sans)
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['mathtext.fontset'] = 'stix'  # Use STIX font for mathematical expressions

# Define datasets and methods
datasets = ['mESC', 'hESC', 'hU2OS', 'mESCQ']
methods = ['Seurat', 'ccRemover', 'HarmoCycle']

# SI score values for each method across datasets
data = {
    'Seurat': [-0.01, 0.08, 0.02, -0.01],
    'ccRemover': [-0.01, 0.222, -0.03, 0.04],
    'HarmoCycle': [-0.05, -0.091, -0.04, -0.183]
}

# Custom color palette for methods
colors = ['#9FAFCB', '#E19A90', '#B9B1C5']

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Define bar positions
x = np.arange(len(datasets))
width = 0.25

# Plot grouped bar chart
bars = []
for i, method in enumerate(methods):
    bar = ax.bar(
        x + i * width, 
        data[method], 
        width, 
        label=method, 
        color=colors[i], 
        edgecolor='black', 
        linewidth=0.5
    )
    bars.append(bar)
    
    # Add numerical labels above/below each bar
    for j, value in enumerate(data[method]):
        ax.text(
            x[j] + i * width, 
            value + (0.01 if value >= 0 else -0.03), 
            f'{value:.3f}', 
            ha='center', 
            va='bottom' if value >= 0 else 'top', 
            fontsize=9
        )

# Axis labels and title
ax.set_xlabel('Dataset', fontsize=12)
ax.set_ylabel('SI Score', fontsize=12)
ax.set_title('Comparison of Cell Cycle Scoring Methods', fontsize=14)

# Set x-axis ticks and labels
ax.set_xticks(x + width)
ax.set_xticklabels(datasets)

# Add legend
ax.legend()

# Add horizontal zero line for reference
ax.axhline(y=0, color='black', linestyle='-', alpha=0.5, linewidth=0.8)

# Adjust y-axis limits (extra space for negative values)
ax.set_ylim(-0.25, 0.3)

# Add gridlines for readability
ax.grid(axis='y', linestyle='--', alpha=0.3)

# Optimize layout
plt.tight_layout()

# ==============================
# Save figure
# ==============================
output_dir = './FigurePlot/CellCycleEffect'
os.makedirs(output_dir, exist_ok=True)

# Save figure as PDF
output_path = os.path.join(output_dir, 'CellCycle_EffectRemove_scoring_comparison.png')
plt.savefig(output_path, format='png', bbox_inches='tight')
print(f"Figure saved as png at: {output_path}")

# Display figure
plt.show()
```

```python
import scanpy as sc
import anndata as ad
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors

# Set plotting parameters
sc.settings.set_figure_params(dpi=80, facecolor='white')
plt.rcParams['font.family'] = 'DejaVu Sans'

# Define custom color palette
color_palette = ['#9FAFCB', '#E19A90', '#B9B1C5', '#A3D9B1']  # First three are given colors, fourth is light green

# Assume four datasets already exist:
# removal_adata1, removal_adata2, removal_adata3, removal_adata4

# 1. Retain only common genes across datasets
common_genes = set(removal_adata1.var_names)
for adata in [removal_adata2, removal_adata3, removal_adata4]:
    common_genes = common_genes.intersection(set(adata.var_names))

common_genes = list(common_genes)
print(f"Number of common genes: {len(common_genes)}")

# Filter each dataset to keep only common genes
removal_adata1 = removal_adata1[:, common_genes].copy()
removal_adata2 = removal_adata2[:, common_genes].copy()
removal_adata3 = removal_adata3[:, common_genes].copy()
removal_adata4 = removal_adata4[:, common_genes].copy()

# 2. Add dataset labels
removal_adata1.obs['dataset'] = 'mESC'
removal_adata2.obs['dataset'] = 'hESC'
removal_adata3.obs['dataset'] = 'hU2OS'
removal_adata4.obs['dataset'] = 'mESCQ'

# 3. Concatenate datasets into one AnnData object
combined_adata = ad.concat(
    [removal_adata1, removal_adata2, removal_adata3, removal_adata4],
    join='outer',
    label='dataset',
    keys=['mESC', 'hESC', 'hU2OS', 'mESCQ'],
    index_unique='-'
)

# Check merged dataset
print(f"Shape after merging: {combined_adata.shape}")
print(f"Dataset distribution:\n{combined_adata.obs['dataset'].value_counts()}")

# 4. PCA analysis
sc.pp.scale(combined_adata, max_value=10)
sc.tl.pca(combined_adata, n_comps=3, svd_solver='arpack')

# Plot PCA results
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# PCA scatter plot (PC1 vs PC2)
sc.pl.pca(combined_adata, color='dataset', ax=axes[0], show=False, palette=color_palette)
axes[0].set_title('PCA: PC1 vs PC2')

# PCA variance explained
pca_variance = combined_adata.uns['pca']['variance_ratio']
axes[1].bar(range(1, 4), pca_variance[:3], color=color_palette[0])
axes[1].set_xlabel('Principal Components')
axes[1].set_ylabel('Variance Ratio')
axes[1].set_title('Variance Explained by First 3 PCs')
axes[1].set_xticks([1, 2, 3])

plt.tight_layout()
plt.savefig('/root/Cycle/Github/FigurePlot/CellCycleEffect/merge_dataset_pca_analysis.png', bbox_inches='tight')
plt.show()

# 5. UMAP analysis
# Build neighbor graph based on PCA
sc.pp.neighbors(combined_adata, n_neighbors=15, n_pcs=3)

# Compute UMAP
sc.tl.umap(combined_adata)

fig, ax = plt.subplots(1, 1, figsize=(8, 6))
sc.pl.umap(combined_adata, color='dataset', ax=ax, show=False, palette=color_palette)
ax.set_title('UMAP Projection by Dataset')
plt.tight_layout()
plt.savefig('/root/Cycle/Github/FigurePlot/CellCycleEffect/merge_dataset_umap_analysis_stage.png', bbox_inches='tight')
plt.show()

# Check dataset labels
print("Dataset categories:", combined_adata.obs['dataset'].astype('category').cat.categories)
print("Number of datasets:", len(combined_adata.obs['dataset'].astype('category').cat.categories))

# Ensure color palette length matches dataset categories
if len(color_palette) < len(combined_adata.obs['dataset'].astype('category').cat.categories):
    print("Warning: Number of colors is fewer than number of datasets. Extend palette or use default.")

# More stable plotting method
fig, ax = plt.subplots(1, 1, figsize=(8, 6))

# Explicitly set color mapping
import matplotlib.colors as mcolors
from matplotlib.cm import get_cmap

cmap = mcolors.ListedColormap(color_palette)
norm = mcolors.BoundaryNorm(range(len(color_palette)+1), cmap.N)

# Scatter plot for UMAP
scatter = ax.scatter(
    combined_adata.obsm['X_umap'][:, 0],
    combined_adata.obsm['X_umap'][:, 1],
    c=combined_adata.obs['dataset'].astype('category').cat.codes,
    cmap=cmap, norm=norm, s=20, alpha=1
)

# Create legend
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w',
           markerfacecolor=color_palette[i], markersize=10,
           label=dataset)
    for i, dataset in enumerate(combined_adata.obs['dataset'].astype('category').cat.categories)
]

ax.legend(handles=legend_elements, loc='best')
ax.set_xlabel('UMAP1')
ax.set_ylabel('UMAP2')
ax.set_title('UMAP Projection by Dataset')

plt.tight_layout()
plt.savefig('/root/Cycle/Github/FigurePlot/CellCycleEffect/merge_dataset_umap_analysis.pdf', bbox_inches='tight', dpi=300)
plt.close()
```
