```python
# ===== Basic libraries =====
import random          # Random number generation
import numpy as np     # Numerical computing
import pandas as pd    # Data manipulation and analysis

# ===== PyTorch =====
import torch           # Core PyTorch library
import torch.nn as nn  # Neural network layers
import torch.optim as optim  # Optimization algorithms
from torch.utils.data import DataLoader  # Dataset and data loading utilities

# ===== Machine learning (scikit-learn) =====
from sklearn.metrics import classification_report  # Model evaluation metrics
from sklearn.preprocessing import StandardScaler, LabelEncoder  # Feature scaling and label encoding
from sklearn.mixture import GaussianMixture  # Gaussian Mixture Models for clustering

# ===== Single-cell analysis =====
import scanpy as sc    # Single-cell RNA-seq analysis toolkit
import anndata as ad   # Annotated data structure for single-cell data

# ===== Visualization =====
import matplotlib.pyplot as plt  # Plotting library
import seaborn as sns            # Statistical data visualization
# ==============================
# Import required packages
# ==============================
import os
import sys
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# ==============================
# Set working directory
# ==============================
os.chdir("/root/Cycle/Github")
print("Current working directory:", os.getcwd())

# ==============================
# Add HarmoCycle module path
# ==============================
sys.path.append("./HarmoCycle/")

# Import custom functions
from HarmoCycle import *
from Utils import *



def run_analysis(SEED):
    random.seed(SEED)

    # Set seed for NumPy
    np.random.seed(SEED)

    # If your code or dependencies (e.g. HarmoCycle) use PyTorch, set the seed there as well
    # This is crucial for reproducibility in neural network initialization
    torch.manual_seed(SEED)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(SEED)
        # For complete reproducibility on CUDA, you may need the following settings
        # Note: this may reduce performance
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

    print(f"Random seeds set to {SEED} for reproducibility.")
    
    
    # Load reference datasets
    adata1 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Buettner.h5ad')
    adata2 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Sasagawa.h5ad')
    adata3 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Mahdessian.h5ad')
    adata4 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Leng.h5ad')

    # Normalize and log-transform each dataset
    sc.pp.normalize_total(adata1, target_sum=1e4)
    sc.pp.log1p(adata1)
    sc.pp.normalize_total(adata2, target_sum=1e4)
    sc.pp.log1p(adata2)
    sc.pp.normalize_total(adata3, target_sum=1e4)   
    sc.pp.log1p(adata3)
    sc.pp.normalize_total(adata4, target_sum=1e4)
    sc.pp.log1p(adata4)
    
    # Shuffle cells in adata1
    adata1 = adata1[np.random.permutation(adata1.n_obs), :]
    _, _, adata_res1= run_pipeline(adata1, epochs=1000, batch_size=2048, top_n_genes=5000, device='cuda:1')
    adata1 = adata1[adata1.obs['pseudo_angle'].sort_values().index, :]

    # Evaluate correlation score
    corr_score_perfect1 = calculate_correlation_score(adata1.obs.stage)
    print(corr_score_perfect1)
    
    # Construct periodic matrix and adjust PCA angle
    periodic_adata = get_periodic_matrix(adata1.copy(), top_k=10, use_highly_variable=False)
    adjust_adata1 = calculate_pca_angle(
        periodic_adata,
        use_rep='X',
        pca_key_added='X_pca_adjust',
        angle_key_added='pca_adjust_angle',
        plot=False
        )
    adjust_adata1 = adjust_adata1[adjust_adata1.obs.sort_values(by='pca_adjust_angle').index, :]
    corr_adjust_score_perfect1 = calculate_correlation_score(adjust_adata1.obs.stage)
    print(corr_adjust_score_perfect1)

    # Build reference dataset dictionary
    reference_dataset = {'ref_dataset1': adata2,
                        'ref_dataset2': adata3,
                        'ref_dataset3': adata4
                        }

    reference_res =  create_reference_gene_sets(reference_dataset)
    adata_adjust = calculate_gene_oscillation_properties_v2(adata1)
    cos_adata = reconstruct_cosine_expression(
        adata_adjust,
        min_frequency=1,
        max_frequency=2,
        min_amplitude=0.01)
    
    # Score cell cycle phases using multiple references
    score_multiple_g1_cell_cycle(
        adata=cos_adata,
        s_genes_list=[reference_res[t]['S'] for t in list(reference_dataset.keys())],
        g2m_genes_list=[reference_res[t]['G2M'] for t in list(reference_dataset.keys())],
        g1_genes_list=[reference_res[t]['G1'] for t in list(reference_dataset.keys())],
        prefix='ref'
    )

    # 1. Standardize cell cycle scores
    cycle_features = [
        'G1_score_ref1', 'S_score_ref1', 'G2M_score_ref1',
        'G1_score_ref2', 'S_score_ref2', 'G2M_score_ref2',
        'G1_score_ref3', 'S_score_ref3', 'G2M_score_ref3',
    ]

    # Only standardize non-missing values
    valid_mask = cos_adata.obs[cycle_features].notnull().all(axis=1)
    scaler = StandardScaler()
    scaled_values = scaler.fit_transform(cos_adata.obs.loc[valid_mask, cycle_features])

    # Store standardized values with '_scaled' suffix
    for i, col in enumerate(cycle_features):
        cos_adata.obs.loc[valid_mask, f'{col}_scaled'] = scaled_values[:, i]

    # 2. Predict phases for each reference
    phases = ['G1', 'S', 'G2M']
    refs = ['ref1', 'ref2', 'ref3']

    for ref in refs:
        # Predicted phase is the column with the maximum score
        cos_adata.obs[f'predicted_phase_{ref}'] = cos_adata.obs[
            [f'{phase}_score_{ref}' for phase in phases]
        ].idxmax(axis=1).str.replace('_score_' + ref, '')

    # 3. Create consensus prediction
    pred_cols = [f'predicted_phase_{ref}' for ref in refs]
    cos_adata.obs['predicted_cell_cycle_consensus'] = cos_adata.obs[pred_cols].mode(axis=1)[0]

    # 4. Filter for consistent predictions across all references
    phase_cols = ['predicted_phase_ref1', 'predicted_phase_ref2', 'predicted_phase_ref3']
    consistent_cells_mask = cos_adata.obs[phase_cols].nunique(axis=1) == 1
    cos_adata.obs['passed_consistency_filter'] = consistent_cells_mask

    # Compute average standardized score for each phase
    phases = ['G1', 'S', 'G2M']
    for phase in phases:
        cols = [f'{phase}_score_ref{i}_scaled' for i in range(1,4)]
        cos_adata.obs[f'{phase}_score_final'] = cos_adata.obs[cols].mean(axis=1)

    # Predict final cell cycle phase
    cos_adata.obs['predicted_phase_final'] = cos_adata.obs[
        ['G1_score_final', 'S_score_final', 'G2M_score_final']
    ].idxmax(axis=1).str.replace('_score_final', '')

    # Scale data (optional if not done earlier)
    sc.pp.scale(cos_adata, max_value=10)

    # Compute PCA with 10 components
    sc.tl.pca(cos_adata, svd_solver='arpack', n_comps=10)

    # Plot PCA colored by stage and scores
    sc.pl.pca(cos_adata, color=['stage', 'G1_score_final', 'S_score_final', 'G2M_score_final'])

    # Corrected scores
    cos_adata.obs['G1_corrected'] = 2*cos_adata.obs['G1_score_final'] - cos_adata.obs['S_score_final'] - cos_adata.obs['G2M_score_final']
    cos_adata.obs['S_corrected'] = 2*cos_adata.obs['S_score_final'] - cos_adata.obs['G1_score_final'] - cos_adata.obs['G2M_score_final']
    cos_adata.obs['G2M_corrected'] = 2*cos_adata.obs['G2M_score_final'] - cos_adata.obs['G1_score_final'] - cos_adata.obs['S_score_final']
    sc.pl.pca(cos_adata, color=['stage', 'S_score_final', 'G1_score_final', 'G2M_score_final', 'G1_corrected', 'S_corrected', 'G2M_corrected'], title='Stage')
    
        # Prepare features for clustering using corrected scores
    features_for_clustering = cos_adata.obs[['G1_corrected', 'S_corrected', 'G2M_corrected']].copy()

    # Check for NaN values, since GMM cannot handle NaN
    if features_for_clustering.isnull().values.any():
        print("Warning: NaN values detected in clustering features. Filling with 0. Consider a better imputation strategy.")
        features_for_clustering.fillna(0, inplace=True)  # Alternatively, use mean or other imputation methods

    # Initialize and fit Gaussian Mixture Model (GMM)
    n_clusters = 10  # Number of clusters (can be tuned)
    gmm = GaussianMixture(n_components=n_clusters, random_state=42, covariance_type='full')
    # covariance_type options: 'full', 'tied', 'diag', 'spherical'

    # Fit the model and predict cluster labels
    cluster_labels = gmm.fit_predict(features_for_clustering)

    # Add clustering results back to AnnData object
    # Convert numeric labels to categorical for better handling in Scanpy
    cos_adata.obs['GMM_clusters'] = pd.Categorical(cluster_labels)

    print(cos_adata.obs['GMM_clusters'].value_counts())

    # Visualize clustering results (recommended)
    # Use PCA plot with cluster labels
    if 'X_pca' not in cos_adata.obsm:
        print("Computing PCA for visualization...")
        sc.tl.pca(cos_adata, svd_solver='arpack')  # Use arpack for reproducibility

    sc.pl.pca(
        cos_adata,
        color=['stage', 'GMM_clusters',
               'G1_corrected', 'S_corrected', 'G2M_corrected'],
        title='PCA with True Stage and GMM Clusters (from corrected scores)',
        ncols=3,  # Adjust number of columns for better layout
    )

    # Ensure GMM clustering and corrected scores exist
    assert 'GMM_clusters' in cos_adata.obs, "Run GMM clustering first"
    required_scores = ['G1_corrected', 'S_corrected', 'G2M_corrected']
    assert all(score in cos_adata.obs for score in required_scores), "Missing corrected score columns"

    # Assign anchor cells based on specificity
    cos_adata = assign_anchor_cells(
        cos_adata,
        specificity_threshold=0.6,
        max_clusters_per_phase=1  # Limit to max 1 cluster per phase
    )

    # Split into source (domain1) and target (domain2)
    domain1 = cos_adata[cos_adata.obs['phase_anchor'] != 'None'].copy()
    domain2 = cos_adata[cos_adata.obs['phase_anchor'] == 'None'].copy()
    domain1 = periodic_adata[domain1.obs.index, ].copy()
    domain2 = periodic_adata[domain2.obs.index, ].copy()

    domain1.obs = cos_adata[domain1.obs.index, ].obs.copy()
    domain2.obs = cos_adata[domain2.obs.index, ].obs.copy()

    # Identify highly variable genes
    sc.pp.highly_variable_genes(periodic_adata, flavor='seurat', n_top_genes=1000)

    # Print number of highly variable genes
    print(f"Number of highly variable genes: {periodic_adata.var['highly_variable'].sum()}")

    # Extract top highly variable genes
    hvg_genes = periodic_adata.var[periodic_adata.var['highly_variable']].index.tolist()
    print(f"Top 10 highly variable genes: {hvg_genes[:10]}")

    # Subset domain1 and domain2 to highly variable genes
    domain1 = domain1[:, hvg_genes]
    domain2 = domain2[:, hvg_genes]

        # --- 1. Data Preparation ---
    # Ensure .X is a dense numpy array for PyTorch
    if hasattr(domain1.X, "toarray"):
        source_X = domain1.X.toarray().astype(np.float32)
    else:
        source_X = domain1.X.astype(np.float32)

    if hasattr(domain2.X, "toarray"):
        target_X = domain2.X.toarray().astype(np.float32)
    else:
        target_X = domain2.X.astype(np.float32)

    # Label encoding for source domain
    label_encoder = LabelEncoder()
    source_y_str = domain1.obs['phase_anchor'].astype(str)
    source_y = label_encoder.fit_transform(source_y_str)
    num_classes = len(label_encoder.classes_)
    print(f"Number of classes: {num_classes}")
    print(f"Class mapping: {dict(zip(label_encoder.classes_, range(num_classes)))}")


    BATCH_SIZE = 5  # Can be tuned

    source_dataset = AnnDataset(source_X, source_y)
    target_dataset = AnnDataset(target_X)  # No labels for target

    source_loader = DataLoader(source_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
    target_loader = DataLoader(target_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)

    # --- 3. Loss Functions and Optimizer ---
    device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    input_dim = source_X.shape[1]
    model = DANN(input_dim=input_dim, num_classes=num_classes).to(device)

    optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
    criterion_label = nn.CrossEntropyLoss()
    criterion_domain = nn.BCEWithLogitsLoss()  # For binary domain classification

    # --- 4. Training Loop ---
    N_EPOCHS = 100  # Can be tuned
    lambda_domain_initial = 0.1  # Initial weight for domain loss
    adapt_grl_alpha = True  # Whether to adapt GRL alpha during training

    print_every = 10

    for epoch in range(N_EPOCHS):
        model.train()
        len_dataloader = min(len(source_loader), len(target_loader))
        iter_source = iter(source_loader)
        iter_target = iter(target_loader)

        total_label_loss = 0
        total_domain_loss = 0
        total_source_acc = 0

        # Adapt GRL alpha (schedule increases domain adaptation importance over time)
        if adapt_grl_alpha:
            p = float(epoch) / N_EPOCHS
            current_alpha = 2. / (1. + np.exp(-10 * p)) - 1
            model.grl.set_alpha(current_alpha)
            current_lambda_domain = 1.0
        else:
            current_lambda_domain = lambda_domain_initial

        for i in range(len_dataloader):
            try:
                source_data, source_labels = next(iter_source)
            except StopIteration:
                iter_source = iter(source_loader)
                source_data, source_labels = next(iter_source)

            try:
                target_data, _ = next(iter_target)
            except StopIteration:
                iter_target = iter(target_loader)
                target_data, _ = next(iter_target)

            source_data, source_labels = source_data.to(device), source_labels.to(device)
            target_data = target_data.to(device)

            optimizer.zero_grad()

            # Source domain forward pass
            label_preds_source, domain_preds_source_logits = model(source_data)
            loss_label = criterion_label(label_preds_source, source_labels)

            # Domain loss for source (label 0)
            domain_labels_source = torch.zeros(source_data.size(0), 1).to(device)
            loss_domain_source = criterion_domain(domain_preds_source_logits, domain_labels_source)

            # Target domain forward pass (only domain prediction used)
            _, domain_preds_target_logits = model(target_data)
            domain_labels_target = torch.ones(target_data.size(0), 1).to(device)
            loss_domain_target = criterion_domain(domain_preds_target_logits, domain_labels_target)

            # Total domain loss
            loss_domain = loss_domain_source + loss_domain_target

            # Total loss
            total_loss = loss_label + current_lambda_domain * loss_domain

            total_loss.backward()
            optimizer.step()

            total_label_loss += loss_label.item()
            total_domain_loss += loss_domain.item()

            # Source accuracy
            _, predicted_labels_source = torch.max(label_preds_source, 1)
            total_source_acc += (predicted_labels_source == source_labels).sum().item()

        avg_label_loss = total_label_loss / len_dataloader
        avg_domain_loss = total_domain_loss / len_dataloader
        avg_source_acc = total_source_acc / (len_dataloader * BATCH_SIZE)

        if (epoch + 1) % print_every == 0 or epoch == 0:
            print(f"Epoch [{epoch+1}/{N_EPOCHS}], Alpha: {model.grl.alpha:.4f}, "
                  f"Label Loss: {avg_label_loss:.4f}, Domain Loss: {avg_domain_loss:.4f}, "
                  f"Source Acc: {avg_source_acc:.4f}")
            # --- 5. Prediction on Target Domain ---
    model.eval()
    target_predictions_list = []
    with torch.no_grad():
        # Re-create target loader without shuffling for ordered predictions
        target_loader_eval = DataLoader(target_dataset, batch_size=BATCH_SIZE, shuffle=False)
        for target_data, _ in target_loader_eval:
            target_data = target_data.to(device)
            output_target = model(target_data, mode='target_predict')
            _, predicted_target = torch.max(output_target, 1)
            target_predictions_list.extend(predicted_target.cpu().numpy())

    # Convert numerical predictions back to original string labels
    predicted_phase_labels = label_encoder.inverse_transform(target_predictions_list)

    # Add predictions to domain2 AnnData object
    # Ensure the length matches. If drop_last=True in DataLoader, some cells might be missed.
    if len(predicted_phase_labels) == domain2.n_obs:
        domain2.obs['predicted_phase_dann'] = predicted_phase_labels
        print("\nPredictions added to domain2.obs['predicted_phase_dann']")
        print(domain2.obs['predicted_phase_dann'].value_counts())
    else:
        print(f"\nWarning: Number of predictions ({len(predicted_phase_labels)}) "
              f"does not match number of target cells ({domain2.n_obs}). "
              "This may happen if drop_last=True was used in target_loader_eval "
              "or if BATCH_SIZE does not divide target_X.shape[0] evenly. "
              "Predictions not added to AnnData.")

    model.eval()
    with torch.no_grad():
        all_X = np.vstack([source_X, target_X])

        # Process in batches if memory is limited
        all_features_list = []
        temp_dataset = AnnDataset(all_X)
        temp_loader = DataLoader(temp_dataset, batch_size=BATCH_SIZE*2, shuffle=False)
        for batch_data, _ in temp_loader:
            batch_data = batch_data.to(device)
            features = model(batch_data, mode='features_only')  # Feature extraction mode
            all_features_list.append(features.cpu().numpy())

        all_features = np.concatenate(all_features_list, axis=0)

    # Create a temporary AnnData for visualization
    adata_for_viz = ad.AnnData(all_features)
    adata_for_viz.obs['domain'] = ['source'] * len(source_X) + ['target'] * len(target_X)
    source_labels_str = domain1.obs['phase_anchor'].tolist()

    # Handle cases where predicted labels are shorter than domain2.n_obs
    target_pred_labels_full = ['Unknown'] * domain2.n_obs
    target_pred_labels_full[:len(predicted_phase_labels)] = predicted_phase_labels
    adata_for_viz.obs['phase'] = source_labels_str + target_pred_labels_full
    adata_for_viz.obs['phase_numeric'] = label_encoder.transform(source_labels_str).tolist() + target_predictions_list[:len(target_pred_labels_full)]

    # Align labels for evaluation
    domain1.obs[['stage', 'phase_anchor']]
    domain2.obs[['stage', 'predicted_phase_dann']]
    domain2.obs.rename(columns={'predicted_phase_dann': 'test'}, inplace=True)
    domain1.obs.rename(columns={'phase_anchor': 'test'}, inplace=True)

    # Replace numeric stage values with string labels
    domain1.obs['stage'] = domain1.obs['stage'].replace({0: 'G1', 1: 'S', 2: 'G2M'})
    domain2.obs['stage'] = domain2.obs['stage'].replace({0: 'G1', 1: 'S', 2: 'G2M'})

    # Generate classification report
    true_labels = pd.concat([domain1.obs['stage'], domain2.obs['stage']])
    predicted_labels = pd.concat([domain1.obs['test'], domain2.obs['test']])

    report_dict = classification_report(true_labels, predicted_labels, output_dict=False)
    return report_dict, domain1, domain2


if __name__ == "__main__":
    # Example seeds
    res1, dataset1, dataset2 = run_analysis(SEED=79710)
```

```python
# ===== Basic libraries =====
import random          # Random number generation
import numpy as np     # Numerical computing
import pandas as pd    # Data manipulation and analysis

# ===== PyTorch =====
import torch           # Core PyTorch library
import torch.nn as nn  # Neural network layers
import torch.optim as optim  # Optimization algorithms
from torch.utils.data import DataLoader  # Dataset and data loading utilities

# ===== Machine learning (scikit-learn) =====
from sklearn.metrics import classification_report  # Model evaluation metrics
from sklearn.preprocessing import StandardScaler, LabelEncoder  # Feature scaling and label encoding
from sklearn.mixture import GaussianMixture  # Gaussian Mixture Models for clustering

# ===== Single-cell analysis =====
import scanpy as sc    # Single-cell RNA-seq analysis toolkit
import anndata as ad   # Annotated data structure for single-cell data

# ===== Visualization =====
import matplotlib.pyplot as plt  # Plotting library
import seaborn as sns            # Statistical data visualization
# ==============================
# Import required packages
# ==============================
import os
import sys
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# ==============================
# Set working directory
# ==============================
os.chdir("/root/Cycle/Github")
print("Current working directory:", os.getcwd())

# ==============================
# Add HarmoCycle module path
# ==============================
sys.path.append("./HarmoCycle/")

# Import custom functions
from HarmoCycle import *
from Utils import *



def run_analysis(SEED):
    random.seed(SEED)

    # Set seed for NumPy
    np.random.seed(SEED)

    # If your code or dependencies (e.g. HarmoCycle) use PyTorch, set the seed there as well
    # This is crucial for reproducibility in neural network initialization
    torch.manual_seed(SEED)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(SEED)
        # For complete reproducibility on CUDA, you may need the following settings
        # Note: this may reduce performance
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

    print(f"Random seeds set to {SEED} for reproducibility.")
    
    
    # Load reference datasets
    adata4 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Buettner.h5ad')
    adata3 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Sasagawa.h5ad')
    adata2 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Mahdessian.h5ad')
    adata1 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Leng.h5ad')

    # Normalize and log-transform each dataset
    sc.pp.normalize_total(adata1, target_sum=1e4)
    sc.pp.log1p(adata1)
    sc.pp.normalize_total(adata2, target_sum=1e4)
    sc.pp.log1p(adata2)
    sc.pp.normalize_total(adata3, target_sum=1e4)   
    sc.pp.log1p(adata3)
    sc.pp.normalize_total(adata4, target_sum=1e4)
    sc.pp.log1p(adata4)
    
    # Shuffle cells in adata1
    adata1 = adata1[np.random.permutation(adata1.n_obs), :]
    _, _, adata_res1= run_pipeline(adata1, epochs=1000, batch_size=2048, top_n_genes=5000, device='cuda:1')
    adata1 = adata1[adata1.obs['pseudo_angle'].sort_values().index, :]

    # Evaluate correlation score
    corr_score_perfect1 = calculate_correlation_score(adata1.obs.stage)
    print(corr_score_perfect1)
    
    # Construct periodic matrix and adjust PCA angle
    periodic_adata = get_periodic_matrix(adata1.copy(), top_k=10, use_highly_variable=False)
    adjust_adata1 = calculate_pca_angle(
        periodic_adata,
        use_rep='X',
        pca_key_added='X_pca_adjust',
        angle_key_added='pca_adjust_angle',
        plot=False
        )
    adjust_adata1 = adjust_adata1[adjust_adata1.obs.sort_values(by='pca_adjust_angle').index, :]
    corr_adjust_score_perfect1 = calculate_correlation_score(adjust_adata1.obs.stage)
    print(corr_adjust_score_perfect1)

    # Build reference dataset dictionary
    reference_dataset = {'ref_dataset1': adata2,
                        'ref_dataset2': adata3,
                        'ref_dataset3': adata4
                        }

    reference_res =  create_reference_gene_sets(reference_dataset)
    adata_adjust = calculate_gene_oscillation_properties_v2(adata1)
    cos_adata = reconstruct_cosine_expression(
        adata_adjust,
        min_frequency=1,
        max_frequency=2,
        min_amplitude=0.01)
    
    # Score cell cycle phases using multiple references
    score_multiple_g1_cell_cycle(
        adata=cos_adata,
        s_genes_list=[reference_res[t]['S'] for t in list(reference_dataset.keys())],
        g2m_genes_list=[reference_res[t]['G2M'] for t in list(reference_dataset.keys())],
        g1_genes_list=[reference_res[t]['G1'] for t in list(reference_dataset.keys())],
        prefix='ref'
    )

    # 1. Standardize cell cycle scores
    cycle_features = [
        'G1_score_ref1', 'S_score_ref1', 'G2M_score_ref1',
        'G1_score_ref2', 'S_score_ref2', 'G2M_score_ref2',
        'G1_score_ref3', 'S_score_ref3', 'G2M_score_ref3',
    ]

    # Only standardize non-missing values
    valid_mask = cos_adata.obs[cycle_features].notnull().all(axis=1)
    scaler = StandardScaler()
    scaled_values = scaler.fit_transform(cos_adata.obs.loc[valid_mask, cycle_features])

    # Store standardized values with '_scaled' suffix
    for i, col in enumerate(cycle_features):
        cos_adata.obs.loc[valid_mask, f'{col}_scaled'] = scaled_values[:, i]

    # 2. Predict phases for each reference
    phases = ['G1', 'S', 'G2M']
    refs = ['ref1', 'ref2', 'ref3']

    for ref in refs:
        # Predicted phase is the column with the maximum score
        cos_adata.obs[f'predicted_phase_{ref}'] = cos_adata.obs[
            [f'{phase}_score_{ref}' for phase in phases]
        ].idxmax(axis=1).str.replace('_score_' + ref, '')

    # 3. Create consensus prediction
    pred_cols = [f'predicted_phase_{ref}' for ref in refs]
    cos_adata.obs['predicted_cell_cycle_consensus'] = cos_adata.obs[pred_cols].mode(axis=1)[0]

    # 4. Filter for consistent predictions across all references
    phase_cols = ['predicted_phase_ref1', 'predicted_phase_ref2', 'predicted_phase_ref3']
    consistent_cells_mask = cos_adata.obs[phase_cols].nunique(axis=1) == 1
    cos_adata.obs['passed_consistency_filter'] = consistent_cells_mask

    # Compute average standardized score for each phase
    phases = ['G1', 'S', 'G2M']
    for phase in phases:
        cols = [f'{phase}_score_ref{i}_scaled' for i in range(1,4)]
        cos_adata.obs[f'{phase}_score_final'] = cos_adata.obs[cols].mean(axis=1)

    # Predict final cell cycle phase
    cos_adata.obs['predicted_phase_final'] = cos_adata.obs[
        ['G1_score_final', 'S_score_final', 'G2M_score_final']
    ].idxmax(axis=1).str.replace('_score_final', '')

    # Scale data (optional if not done earlier)
    sc.pp.scale(cos_adata, max_value=10)

    # Compute PCA with 10 components
    sc.tl.pca(cos_adata, svd_solver='arpack', n_comps=10)

    # Plot PCA colored by stage and scores
    sc.pl.pca(cos_adata, color=['stage', 'G1_score_final', 'S_score_final', 'G2M_score_final'])

    # Corrected scores
    cos_adata.obs['G1_corrected'] = 2*cos_adata.obs['G1_score_final'] - cos_adata.obs['S_score_final'] - cos_adata.obs['G2M_score_final']
    cos_adata.obs['S_corrected'] = 2*cos_adata.obs['S_score_final'] - cos_adata.obs['G1_score_final'] - cos_adata.obs['G2M_score_final']
    cos_adata.obs['G2M_corrected'] = 2*cos_adata.obs['G2M_score_final'] - cos_adata.obs['G1_score_final'] - cos_adata.obs['S_score_final']
    sc.pl.pca(cos_adata, color=['stage', 'S_score_final', 'G1_score_final', 'G2M_score_final', 'G1_corrected', 'S_corrected', 'G2M_corrected'], title='Stage')
    
        # Prepare features for clustering using corrected scores
    features_for_clustering = cos_adata.obs[['G1_corrected', 'S_corrected', 'G2M_corrected']].copy()

    # Check for NaN values, since GMM cannot handle NaN
    if features_for_clustering.isnull().values.any():
        print("Warning: NaN values detected in clustering features. Filling with 0. Consider a better imputation strategy.")
        features_for_clustering.fillna(0, inplace=True)  # Alternatively, use mean or other imputation methods

    # Initialize and fit Gaussian Mixture Model (GMM)
    n_clusters = 10  # Number of clusters (can be tuned)
    gmm = GaussianMixture(n_components=n_clusters, random_state=42, covariance_type='full')
    # covariance_type options: 'full', 'tied', 'diag', 'spherical'

    # Fit the model and predict cluster labels
    cluster_labels = gmm.fit_predict(features_for_clustering)

    # Add clustering results back to AnnData object
    # Convert numeric labels to categorical for better handling in Scanpy
    cos_adata.obs['GMM_clusters'] = pd.Categorical(cluster_labels)

    print(cos_adata.obs['GMM_clusters'].value_counts())

    # Visualize clustering results (recommended)
    # Use PCA plot with cluster labels
    if 'X_pca' not in cos_adata.obsm:
        print("Computing PCA for visualization...")
        sc.tl.pca(cos_adata, svd_solver='arpack')  # Use arpack for reproducibility

    sc.pl.pca(
        cos_adata,
        color=['stage', 'GMM_clusters',
               'G1_corrected', 'S_corrected', 'G2M_corrected'],
        title='PCA with True Stage and GMM Clusters (from corrected scores)',
        ncols=3,  # Adjust number of columns for better layout
    )

    # Ensure GMM clustering and corrected scores exist
    assert 'GMM_clusters' in cos_adata.obs, "Run GMM clustering first"
    required_scores = ['G1_corrected', 'S_corrected', 'G2M_corrected']
    assert all(score in cos_adata.obs for score in required_scores), "Missing corrected score columns"

    # Assign anchor cells based on specificity
    cos_adata = assign_anchor_cells(
        cos_adata,
        specificity_threshold=0.6,
        max_clusters_per_phase=1  # Limit to max 1 cluster per phase
    )

    # Split into source (domain1) and target (domain2)
    domain1 = cos_adata[cos_adata.obs['phase_anchor'] != 'None'].copy()
    domain2 = cos_adata[cos_adata.obs['phase_anchor'] == 'None'].copy()
    domain1 = periodic_adata[domain1.obs.index, ].copy()
    domain2 = periodic_adata[domain2.obs.index, ].copy()

    domain1.obs = cos_adata[domain1.obs.index, ].obs.copy()
    domain2.obs = cos_adata[domain2.obs.index, ].obs.copy()

    # Identify highly variable genes
    sc.pp.highly_variable_genes(periodic_adata, flavor='seurat', n_top_genes=1000)

    # Print number of highly variable genes
    print(f"Number of highly variable genes: {periodic_adata.var['highly_variable'].sum()}")

    # Extract top highly variable genes
    hvg_genes = periodic_adata.var[periodic_adata.var['highly_variable']].index.tolist()
    print(f"Top 10 highly variable genes: {hvg_genes[:10]}")

    # Subset domain1 and domain2 to highly variable genes
    domain1 = domain1[:, hvg_genes]
    domain2 = domain2[:, hvg_genes]

        # --- 1. Data Preparation ---
    # Ensure .X is a dense numpy array for PyTorch
    if hasattr(domain1.X, "toarray"):
        source_X = domain1.X.toarray().astype(np.float32)
    else:
        source_X = domain1.X.astype(np.float32)

    if hasattr(domain2.X, "toarray"):
        target_X = domain2.X.toarray().astype(np.float32)
    else:
        target_X = domain2.X.astype(np.float32)

    # Label encoding for source domain
    label_encoder = LabelEncoder()
    source_y_str = domain1.obs['phase_anchor'].astype(str)
    source_y = label_encoder.fit_transform(source_y_str)
    num_classes = len(label_encoder.classes_)
    print(f"Number of classes: {num_classes}")
    print(f"Class mapping: {dict(zip(label_encoder.classes_, range(num_classes)))}")


    BATCH_SIZE = 5  # Can be tuned

    source_dataset = AnnDataset(source_X, source_y)
    target_dataset = AnnDataset(target_X)  # No labels for target

    source_loader = DataLoader(source_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
    target_loader = DataLoader(target_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)

    # --- 3. Loss Functions and Optimizer ---
    device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    input_dim = source_X.shape[1]
    model = DANN(input_dim=input_dim, num_classes=num_classes).to(device)

    optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
    criterion_label = nn.CrossEntropyLoss()
    criterion_domain = nn.BCEWithLogitsLoss()  # For binary domain classification

    # --- 4. Training Loop ---
    N_EPOCHS = 100  # Can be tuned
    lambda_domain_initial = 0.1  # Initial weight for domain loss
    adapt_grl_alpha = True  # Whether to adapt GRL alpha during training

    print_every = 10

    for epoch in range(N_EPOCHS):
        model.train()
        len_dataloader = min(len(source_loader), len(target_loader))
        iter_source = iter(source_loader)
        iter_target = iter(target_loader)

        total_label_loss = 0
        total_domain_loss = 0
        total_source_acc = 0

        # Adapt GRL alpha (schedule increases domain adaptation importance over time)
        if adapt_grl_alpha:
            p = float(epoch) / N_EPOCHS
            current_alpha = 2. / (1. + np.exp(-10 * p)) - 1
            model.grl.set_alpha(current_alpha)
            current_lambda_domain = 1.0
        else:
            current_lambda_domain = lambda_domain_initial

        for i in range(len_dataloader):
            try:
                source_data, source_labels = next(iter_source)
            except StopIteration:
                iter_source = iter(source_loader)
                source_data, source_labels = next(iter_source)

            try:
                target_data, _ = next(iter_target)
            except StopIteration:
                iter_target = iter(target_loader)
                target_data, _ = next(iter_target)

            source_data, source_labels = source_data.to(device), source_labels.to(device)
            target_data = target_data.to(device)

            optimizer.zero_grad()

            # Source domain forward pass
            label_preds_source, domain_preds_source_logits = model(source_data)
            loss_label = criterion_label(label_preds_source, source_labels)

            # Domain loss for source (label 0)
            domain_labels_source = torch.zeros(source_data.size(0), 1).to(device)
            loss_domain_source = criterion_domain(domain_preds_source_logits, domain_labels_source)

            # Target domain forward pass (only domain prediction used)
            _, domain_preds_target_logits = model(target_data)
            domain_labels_target = torch.ones(target_data.size(0), 1).to(device)
            loss_domain_target = criterion_domain(domain_preds_target_logits, domain_labels_target)

            # Total domain loss
            loss_domain = loss_domain_source + loss_domain_target

            # Total loss
            total_loss = loss_label + current_lambda_domain * loss_domain

            total_loss.backward()
            optimizer.step()

            total_label_loss += loss_label.item()
            total_domain_loss += loss_domain.item()

            # Source accuracy
            _, predicted_labels_source = torch.max(label_preds_source, 1)
            total_source_acc += (predicted_labels_source == source_labels).sum().item()

        avg_label_loss = total_label_loss / len_dataloader
        avg_domain_loss = total_domain_loss / len_dataloader
        avg_source_acc = total_source_acc / (len_dataloader * BATCH_SIZE)

        if (epoch + 1) % print_every == 0 or epoch == 0:
            print(f"Epoch [{epoch+1}/{N_EPOCHS}], Alpha: {model.grl.alpha:.4f}, "
                  f"Label Loss: {avg_label_loss:.4f}, Domain Loss: {avg_domain_loss:.4f}, "
                  f"Source Acc: {avg_source_acc:.4f}")
            # --- 5. Prediction on Target Domain ---
    model.eval()
    target_predictions_list = []
    with torch.no_grad():
        # Re-create target loader without shuffling for ordered predictions
        target_loader_eval = DataLoader(target_dataset, batch_size=BATCH_SIZE, shuffle=False)
        for target_data, _ in target_loader_eval:
            target_data = target_data.to(device)
            output_target = model(target_data, mode='target_predict')
            _, predicted_target = torch.max(output_target, 1)
            target_predictions_list.extend(predicted_target.cpu().numpy())

    # Convert numerical predictions back to original string labels
    predicted_phase_labels = label_encoder.inverse_transform(target_predictions_list)

    # Add predictions to domain2 AnnData object
    # Ensure the length matches. If drop_last=True in DataLoader, some cells might be missed.
    if len(predicted_phase_labels) == domain2.n_obs:
        domain2.obs['predicted_phase_dann'] = predicted_phase_labels
        print("\nPredictions added to domain2.obs['predicted_phase_dann']")
        print(domain2.obs['predicted_phase_dann'].value_counts())
    else:
        print(f"\nWarning: Number of predictions ({len(predicted_phase_labels)}) "
              f"does not match number of target cells ({domain2.n_obs}). "
              "This may happen if drop_last=True was used in target_loader_eval "
              "or if BATCH_SIZE does not divide target_X.shape[0] evenly. "
              "Predictions not added to AnnData.")

    model.eval()
    with torch.no_grad():
        all_X = np.vstack([source_X, target_X])

        # Process in batches if memory is limited
        all_features_list = []
        temp_dataset = AnnDataset(all_X)
        temp_loader = DataLoader(temp_dataset, batch_size=BATCH_SIZE*2, shuffle=False)
        for batch_data, _ in temp_loader:
            batch_data = batch_data.to(device)
            features = model(batch_data, mode='features_only')  # Feature extraction mode
            all_features_list.append(features.cpu().numpy())

        all_features = np.concatenate(all_features_list, axis=0)

    # Create a temporary AnnData for visualization
    adata_for_viz = ad.AnnData(all_features)
    adata_for_viz.obs['domain'] = ['source'] * len(source_X) + ['target'] * len(target_X)
    source_labels_str = domain1.obs['phase_anchor'].tolist()

    # Handle cases where predicted labels are shorter than domain2.n_obs
    target_pred_labels_full = ['Unknown'] * domain2.n_obs
    target_pred_labels_full[:len(predicted_phase_labels)] = predicted_phase_labels
    adata_for_viz.obs['phase'] = source_labels_str + target_pred_labels_full
    adata_for_viz.obs['phase_numeric'] = label_encoder.transform(source_labels_str).tolist() + target_predictions_list[:len(target_pred_labels_full)]

    # Align labels for evaluation
    domain1.obs[['stage', 'phase_anchor']]
    domain2.obs[['stage', 'predicted_phase_dann']]
    domain2.obs.rename(columns={'predicted_phase_dann': 'test'}, inplace=True)
    domain1.obs.rename(columns={'phase_anchor': 'test'}, inplace=True)

    # Replace numeric stage values with string labels
    domain1.obs['stage'] = domain1.obs['stage'].replace({0: 'G1', 1: 'S', 2: 'G2M'})
    domain2.obs['stage'] = domain2.obs['stage'].replace({0: 'G1', 1: 'S', 2: 'G2M'})

    # Generate classification report
    true_labels = pd.concat([domain1.obs['stage'], domain2.obs['stage']])
    predicted_labels = pd.concat([domain1.obs['test'], domain2.obs['test']])

    report_dict = classification_report(true_labels, predicted_labels, output_dict=False)
    return report_dict, domain1, domain2


if __name__ == "__main__":
    # Example seeds
    res2, dataset1, dataset2 = run_analysis(SEED=79710)
```

```python
# ===== Basic libraries =====
import random          # Random number generation
import numpy as np     # Numerical computing
import pandas as pd    # Data manipulation and analysis

# ===== PyTorch =====
import torch           # Core PyTorch library
import torch.nn as nn  # Neural network layers
import torch.optim as optim  # Optimization algorithms
from torch.utils.data import DataLoader  # Dataset and data loading utilities

# ===== Machine learning (scikit-learn) =====
from sklearn.metrics import classification_report  # Model evaluation metrics
from sklearn.preprocessing import StandardScaler, LabelEncoder  # Feature scaling and label encoding
from sklearn.mixture import GaussianMixture  # Gaussian Mixture Models for clustering

# ===== Single-cell analysis =====
import scanpy as sc    # Single-cell RNA-seq analysis toolkit
import anndata as ad   # Annotated data structure for single-cell data

# ===== Visualization =====
import matplotlib.pyplot as plt  # Plotting library
import seaborn as sns            # Statistical data visualization
# ==============================
# Import required packages
# ==============================
import os
import sys
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# ==============================
# Set working directory
# ==============================
os.chdir("/root/Cycle/Github")
print("Current working directory:", os.getcwd())

# ==============================
# Add HarmoCycle module path
# ==============================
sys.path.append("./HarmoCycle/")

# Import custom functions
from HarmoCycle import *
from Utils import *



def run_analysis(SEED):
    random.seed(SEED)

    # Set seed for NumPy
    np.random.seed(SEED)

    # If your code or dependencies (e.g. HarmoCycle) use PyTorch, set the seed there as well
    # This is crucial for reproducibility in neural network initialization
    torch.manual_seed(SEED)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(SEED)
        # For complete reproducibility on CUDA, you may need the following settings
        # Note: this may reduce performance
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

    print(f"Random seeds set to {SEED} for reproducibility.")
    
    
    # Load reference datasets
    adata4 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Buettner.h5ad')
    adata3 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Sasagawa.h5ad')
    adata1 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Mahdessian.h5ad')
    adata2 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Leng.h5ad')

    # Normalize and log-transform each dataset
    sc.pp.normalize_total(adata1, target_sum=1e4)
    sc.pp.log1p(adata1)
    sc.pp.normalize_total(adata2, target_sum=1e4)
    sc.pp.log1p(adata2)
    sc.pp.normalize_total(adata3, target_sum=1e4)   
    sc.pp.log1p(adata3)
    sc.pp.normalize_total(adata4, target_sum=1e4)
    sc.pp.log1p(adata4)
    
    # Shuffle cells in adata1
    adata1 = adata1[np.random.permutation(adata1.n_obs), :]
    _, _, adata_res1= run_pipeline(adata1, epochs=1000, batch_size=2048, top_n_genes=5000, device='cuda:1')
    adata1 = adata1[adata1.obs['pseudo_angle'].sort_values().index, :]

    # Evaluate correlation score
    corr_score_perfect1 = calculate_correlation_score(adata1.obs.stage)
    print(corr_score_perfect1)
    
    # Construct periodic matrix and adjust PCA angle
    periodic_adata = get_periodic_matrix(adata1.copy(), top_k=10, use_highly_variable=False)
    adjust_adata1 = calculate_pca_angle(
        periodic_adata,
        use_rep='X',
        pca_key_added='X_pca_adjust',
        angle_key_added='pca_adjust_angle',
        plot=False
        )
    adjust_adata1 = adjust_adata1[adjust_adata1.obs.sort_values(by='pca_adjust_angle').index, :]
    corr_adjust_score_perfect1 = calculate_correlation_score(adjust_adata1.obs.stage)
    print(corr_adjust_score_perfect1)

    # Build reference dataset dictionary
    reference_dataset = {'ref_dataset1': adata2,
                        'ref_dataset2': adata3,
                        'ref_dataset3': adata4
                        }

    reference_res =  create_reference_gene_sets(reference_dataset)
    adata_adjust = calculate_gene_oscillation_properties_v2(adata1)
    cos_adata = reconstruct_cosine_expression(
        adata_adjust,
        min_frequency=1,
        max_frequency=2,
        min_amplitude=0.01)
    
    # Score cell cycle phases using multiple references
    score_multiple_g1_cell_cycle(
        adata=cos_adata,
        s_genes_list=[reference_res[t]['S'] for t in list(reference_dataset.keys())],
        g2m_genes_list=[reference_res[t]['G2M'] for t in list(reference_dataset.keys())],
        g1_genes_list=[reference_res[t]['G1'] for t in list(reference_dataset.keys())],
        prefix='ref'
    )

    # 1. Standardize cell cycle scores
    cycle_features = [
        'G1_score_ref1', 'S_score_ref1', 'G2M_score_ref1',
        'G1_score_ref2', 'S_score_ref2', 'G2M_score_ref2',
        'G1_score_ref3', 'S_score_ref3', 'G2M_score_ref3',
    ]

    # Only standardize non-missing values
    valid_mask = cos_adata.obs[cycle_features].notnull().all(axis=1)
    scaler = StandardScaler()
    scaled_values = scaler.fit_transform(cos_adata.obs.loc[valid_mask, cycle_features])

    # Store standardized values with '_scaled' suffix
    for i, col in enumerate(cycle_features):
        cos_adata.obs.loc[valid_mask, f'{col}_scaled'] = scaled_values[:, i]

    # 2. Predict phases for each reference
    phases = ['G1', 'S', 'G2M']
    refs = ['ref1', 'ref2', 'ref3']

    for ref in refs:
        # Predicted phase is the column with the maximum score
        cos_adata.obs[f'predicted_phase_{ref}'] = cos_adata.obs[
            [f'{phase}_score_{ref}' for phase in phases]
        ].idxmax(axis=1).str.replace('_score_' + ref, '')

    # 3. Create consensus prediction
    pred_cols = [f'predicted_phase_{ref}' for ref in refs]
    cos_adata.obs['predicted_cell_cycle_consensus'] = cos_adata.obs[pred_cols].mode(axis=1)[0]

    # 4. Filter for consistent predictions across all references
    phase_cols = ['predicted_phase_ref1', 'predicted_phase_ref2', 'predicted_phase_ref3']
    consistent_cells_mask = cos_adata.obs[phase_cols].nunique(axis=1) == 1
    cos_adata.obs['passed_consistency_filter'] = consistent_cells_mask

    # Compute average standardized score for each phase
    phases = ['G1', 'S', 'G2M']
    for phase in phases:
        cols = [f'{phase}_score_ref{i}_scaled' for i in range(1,4)]
        cos_adata.obs[f'{phase}_score_final'] = cos_adata.obs[cols].mean(axis=1)

    # Predict final cell cycle phase
    cos_adata.obs['predicted_phase_final'] = cos_adata.obs[
        ['G1_score_final', 'S_score_final', 'G2M_score_final']
    ].idxmax(axis=1).str.replace('_score_final', '')

    # Scale data (optional if not done earlier)
    sc.pp.scale(cos_adata, max_value=10)

    # Compute PCA with 10 components
    sc.tl.pca(cos_adata, svd_solver='arpack', n_comps=10)

    # Plot PCA colored by stage and scores
    sc.pl.pca(cos_adata, color=['stage', 'G1_score_final', 'S_score_final', 'G2M_score_final'])

    # Corrected scores
    cos_adata.obs['G1_corrected'] = 2*cos_adata.obs['G1_score_final'] - cos_adata.obs['S_score_final'] - cos_adata.obs['G2M_score_final']
    cos_adata.obs['S_corrected'] = 2*cos_adata.obs['S_score_final'] - cos_adata.obs['G1_score_final'] - cos_adata.obs['G2M_score_final']
    cos_adata.obs['G2M_corrected'] = 2*cos_adata.obs['G2M_score_final'] - cos_adata.obs['G1_score_final'] - cos_adata.obs['S_score_final']
    sc.pl.pca(cos_adata, color=['stage', 'S_score_final', 'G1_score_final', 'G2M_score_final', 'G1_corrected', 'S_corrected', 'G2M_corrected'], title='Stage')
    
        # Prepare features for clustering using corrected scores
    features_for_clustering = cos_adata.obs[['G1_corrected', 'S_corrected', 'G2M_corrected']].copy()

    # Check for NaN values, since GMM cannot handle NaN
    if features_for_clustering.isnull().values.any():
        print("Warning: NaN values detected in clustering features. Filling with 0. Consider a better imputation strategy.")
        features_for_clustering.fillna(0, inplace=True)  # Alternatively, use mean or other imputation methods

    # Initialize and fit Gaussian Mixture Model (GMM)
    n_clusters = 10  # Number of clusters (can be tuned)
    gmm = GaussianMixture(n_components=n_clusters, random_state=42, covariance_type='full')
    # covariance_type options: 'full', 'tied', 'diag', 'spherical'

    # Fit the model and predict cluster labels
    cluster_labels = gmm.fit_predict(features_for_clustering)

    # Add clustering results back to AnnData object
    # Convert numeric labels to categorical for better handling in Scanpy
    cos_adata.obs['GMM_clusters'] = pd.Categorical(cluster_labels)

    print(cos_adata.obs['GMM_clusters'].value_counts())

    # Visualize clustering results (recommended)
    # Use PCA plot with cluster labels
    if 'X_pca' not in cos_adata.obsm:
        print("Computing PCA for visualization...")
        sc.tl.pca(cos_adata, svd_solver='arpack')  # Use arpack for reproducibility

    sc.pl.pca(
        cos_adata,
        color=['stage', 'GMM_clusters',
               'G1_corrected', 'S_corrected', 'G2M_corrected'],
        title='PCA with True Stage and GMM Clusters (from corrected scores)',
        ncols=3,  # Adjust number of columns for better layout
    )

    # Ensure GMM clustering and corrected scores exist
    assert 'GMM_clusters' in cos_adata.obs, "Run GMM clustering first"
    required_scores = ['G1_corrected', 'S_corrected', 'G2M_corrected']
    assert all(score in cos_adata.obs for score in required_scores), "Missing corrected score columns"

    # Assign anchor cells based on specificity
    cos_adata = assign_anchor_cells(
        cos_adata,
        specificity_threshold=0.6,
        max_clusters_per_phase=1  # Limit to max 1 cluster per phase
    )

    # Split into source (domain1) and target (domain2)
    domain1 = cos_adata[cos_adata.obs['phase_anchor'] != 'None'].copy()
    domain2 = cos_adata[cos_adata.obs['phase_anchor'] == 'None'].copy()
    domain1 = periodic_adata[domain1.obs.index, ].copy()
    domain2 = periodic_adata[domain2.obs.index, ].copy()

    domain1.obs = cos_adata[domain1.obs.index, ].obs.copy()
    domain2.obs = cos_adata[domain2.obs.index, ].obs.copy()

    # Identify highly variable genes
    sc.pp.highly_variable_genes(periodic_adata, flavor='seurat', n_top_genes=1000)

    # Print number of highly variable genes
    print(f"Number of highly variable genes: {periodic_adata.var['highly_variable'].sum()}")

    # Extract top highly variable genes
    hvg_genes = periodic_adata.var[periodic_adata.var['highly_variable']].index.tolist()
    print(f"Top 10 highly variable genes: {hvg_genes[:10]}")

    # Subset domain1 and domain2 to highly variable genes
    domain1 = domain1[:, hvg_genes]
    domain2 = domain2[:, hvg_genes]

        # --- 1. Data Preparation ---
    # Ensure .X is a dense numpy array for PyTorch
    if hasattr(domain1.X, "toarray"):
        source_X = domain1.X.toarray().astype(np.float32)
    else:
        source_X = domain1.X.astype(np.float32)

    if hasattr(domain2.X, "toarray"):
        target_X = domain2.X.toarray().astype(np.float32)
    else:
        target_X = domain2.X.astype(np.float32)

    # Label encoding for source domain
    label_encoder = LabelEncoder()
    source_y_str = domain1.obs['phase_anchor'].astype(str)
    source_y = label_encoder.fit_transform(source_y_str)
    num_classes = len(label_encoder.classes_)
    print(f"Number of classes: {num_classes}")
    print(f"Class mapping: {dict(zip(label_encoder.classes_, range(num_classes)))}")


    BATCH_SIZE = 5  # Can be tuned

    source_dataset = AnnDataset(source_X, source_y)
    target_dataset = AnnDataset(target_X)  # No labels for target

    source_loader = DataLoader(source_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
    target_loader = DataLoader(target_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)

    # --- 3. Loss Functions and Optimizer ---
    device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    input_dim = source_X.shape[1]
    model = DANN(input_dim=input_dim, num_classes=num_classes).to(device)

    optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
    criterion_label = nn.CrossEntropyLoss()
    criterion_domain = nn.BCEWithLogitsLoss()  # For binary domain classification

    # --- 4. Training Loop ---
    N_EPOCHS = 100  # Can be tuned
    lambda_domain_initial = 0.1  # Initial weight for domain loss
    adapt_grl_alpha = True  # Whether to adapt GRL alpha during training

    print_every = 10

    for epoch in range(N_EPOCHS):
        model.train()
        len_dataloader = min(len(source_loader), len(target_loader))
        iter_source = iter(source_loader)
        iter_target = iter(target_loader)

        total_label_loss = 0
        total_domain_loss = 0
        total_source_acc = 0

        # Adapt GRL alpha (schedule increases domain adaptation importance over time)
        if adapt_grl_alpha:
            p = float(epoch) / N_EPOCHS
            current_alpha = 2. / (1. + np.exp(-10 * p)) - 1
            model.grl.set_alpha(current_alpha)
            current_lambda_domain = 1.0
        else:
            current_lambda_domain = lambda_domain_initial

        for i in range(len_dataloader):
            try:
                source_data, source_labels = next(iter_source)
            except StopIteration:
                iter_source = iter(source_loader)
                source_data, source_labels = next(iter_source)

            try:
                target_data, _ = next(iter_target)
            except StopIteration:
                iter_target = iter(target_loader)
                target_data, _ = next(iter_target)

            source_data, source_labels = source_data.to(device), source_labels.to(device)
            target_data = target_data.to(device)

            optimizer.zero_grad()

            # Source domain forward pass
            label_preds_source, domain_preds_source_logits = model(source_data)
            loss_label = criterion_label(label_preds_source, source_labels)

            # Domain loss for source (label 0)
            domain_labels_source = torch.zeros(source_data.size(0), 1).to(device)
            loss_domain_source = criterion_domain(domain_preds_source_logits, domain_labels_source)

            # Target domain forward pass (only domain prediction used)
            _, domain_preds_target_logits = model(target_data)
            domain_labels_target = torch.ones(target_data.size(0), 1).to(device)
            loss_domain_target = criterion_domain(domain_preds_target_logits, domain_labels_target)

            # Total domain loss
            loss_domain = loss_domain_source + loss_domain_target

            # Total loss
            total_loss = loss_label + current_lambda_domain * loss_domain

            total_loss.backward()
            optimizer.step()

            total_label_loss += loss_label.item()
            total_domain_loss += loss_domain.item()

            # Source accuracy
            _, predicted_labels_source = torch.max(label_preds_source, 1)
            total_source_acc += (predicted_labels_source == source_labels).sum().item()

        avg_label_loss = total_label_loss / len_dataloader
        avg_domain_loss = total_domain_loss / len_dataloader
        avg_source_acc = total_source_acc / (len_dataloader * BATCH_SIZE)

        if (epoch + 1) % print_every == 0 or epoch == 0:
            print(f"Epoch [{epoch+1}/{N_EPOCHS}], Alpha: {model.grl.alpha:.4f}, "
                  f"Label Loss: {avg_label_loss:.4f}, Domain Loss: {avg_domain_loss:.4f}, "
                  f"Source Acc: {avg_source_acc:.4f}")
            # --- 5. Prediction on Target Domain ---
    model.eval()
    target_predictions_list = []
    with torch.no_grad():
        # Re-create target loader without shuffling for ordered predictions
        target_loader_eval = DataLoader(target_dataset, batch_size=BATCH_SIZE, shuffle=False)
        for target_data, _ in target_loader_eval:
            target_data = target_data.to(device)
            output_target = model(target_data, mode='target_predict')
            _, predicted_target = torch.max(output_target, 1)
            target_predictions_list.extend(predicted_target.cpu().numpy())

    # Convert numerical predictions back to original string labels
    predicted_phase_labels = label_encoder.inverse_transform(target_predictions_list)

    # Add predictions to domain2 AnnData object
    # Ensure the length matches. If drop_last=True in DataLoader, some cells might be missed.
    if len(predicted_phase_labels) == domain2.n_obs:
        domain2.obs['predicted_phase_dann'] = predicted_phase_labels
        print("\nPredictions added to domain2.obs['predicted_phase_dann']")
        print(domain2.obs['predicted_phase_dann'].value_counts())
    else:
        print(f"\nWarning: Number of predictions ({len(predicted_phase_labels)}) "
              f"does not match number of target cells ({domain2.n_obs}). "
              "This may happen if drop_last=True was used in target_loader_eval "
              "or if BATCH_SIZE does not divide target_X.shape[0] evenly. "
              "Predictions not added to AnnData.")

    model.eval()
    with torch.no_grad():
        all_X = np.vstack([source_X, target_X])

        # Process in batches if memory is limited
        all_features_list = []
        temp_dataset = AnnDataset(all_X)
        temp_loader = DataLoader(temp_dataset, batch_size=BATCH_SIZE*2, shuffle=False)
        for batch_data, _ in temp_loader:
            batch_data = batch_data.to(device)
            features = model(batch_data, mode='features_only')  # Feature extraction mode
            all_features_list.append(features.cpu().numpy())

        all_features = np.concatenate(all_features_list, axis=0)

    # Create a temporary AnnData for visualization
    adata_for_viz = ad.AnnData(all_features)
    adata_for_viz.obs['domain'] = ['source'] * len(source_X) + ['target'] * len(target_X)
    source_labels_str = domain1.obs['phase_anchor'].tolist()

    # Handle cases where predicted labels are shorter than domain2.n_obs
    target_pred_labels_full = ['Unknown'] * domain2.n_obs
    target_pred_labels_full[:len(predicted_phase_labels)] = predicted_phase_labels
    adata_for_viz.obs['phase'] = source_labels_str + target_pred_labels_full
    adata_for_viz.obs['phase_numeric'] = label_encoder.transform(source_labels_str).tolist() + target_predictions_list[:len(target_pred_labels_full)]

    # Align labels for evaluation
    domain1.obs[['stage', 'phase_anchor']]
    domain2.obs[['stage', 'predicted_phase_dann']]
    domain2.obs.rename(columns={'predicted_phase_dann': 'test'}, inplace=True)
    domain1.obs.rename(columns={'phase_anchor': 'test'}, inplace=True)

    # Replace numeric stage values with string labels
    domain1.obs['stage'] = domain1.obs['stage'].replace({0: 'G1', 1: 'S', 2: 'G2M'})
    domain2.obs['stage'] = domain2.obs['stage'].replace({0: 'G1', 1: 'S', 2: 'G2M'})

    # Generate classification report
    true_labels = pd.concat([domain1.obs['stage'], domain2.obs['stage']])
    predicted_labels = pd.concat([domain1.obs['test'], domain2.obs['test']])

    report_dict = classification_report(true_labels, predicted_labels, output_dict=False)
    return report_dict, domain1, domain2


if __name__ == "__main__":
    # Example seeds
    res3, dataset1, dataset2 = run_analysis(SEED=79710)
```

```python
# ===== Basic libraries =====
import random          # Random number generation
import numpy as np     # Numerical computing
import pandas as pd    # Data manipulation and analysis

# ===== PyTorch =====
import torch           # Core PyTorch library
import torch.nn as nn  # Neural network layers
import torch.optim as optim  # Optimization algorithms
from torch.utils.data import DataLoader  # Dataset and data loading utilities

# ===== Machine learning (scikit-learn) =====
from sklearn.metrics import classification_report  # Model evaluation metrics
from sklearn.preprocessing import StandardScaler, LabelEncoder  # Feature scaling and label encoding
from sklearn.mixture import GaussianMixture  # Gaussian Mixture Models for clustering

# ===== Single-cell analysis =====
import scanpy as sc    # Single-cell RNA-seq analysis toolkit
import anndata as ad   # Annotated data structure for single-cell data

# ===== Visualization =====
import matplotlib.pyplot as plt  # Plotting library
import seaborn as sns            # Statistical data visualization
# ==============================
# Import required packages
# ==============================
import os
import sys
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# ==============================
# Set working directory
# ==============================
os.chdir("/root/Cycle/Github")
print("Current working directory:", os.getcwd())

# ==============================
# Add HarmoCycle module path
# ==============================
sys.path.append("./HarmoCycle/")

# Import custom functions
from HarmoCycle import *
from Utils import *



def run_analysis(SEED):
    random.seed(SEED)

    # Set seed for NumPy
    np.random.seed(SEED)

    # If your code or dependencies (e.g. HarmoCycle) use PyTorch, set the seed there as well
    # This is crucial for reproducibility in neural network initialization
    torch.manual_seed(SEED)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(SEED)
        # For complete reproducibility on CUDA, you may need the following settings
        # Note: this may reduce performance
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

    print(f"Random seeds set to {SEED} for reproducibility.")
    
    
    # Load reference datasets
    adata2 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Buettner.h5ad')
    adata1 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Sasagawa.h5ad')
    adata4 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Mahdessian.h5ad')
    adata3 = sc.read('/remote-home/share/data2/wbl/wbl_cellcycle/ground-truth/ref_dataset_Leng.h5ad')

    # Normalize and log-transform each dataset
    sc.pp.normalize_total(adata1, target_sum=1e4)
    sc.pp.log1p(adata1)
    sc.pp.normalize_total(adata2, target_sum=1e4)
    sc.pp.log1p(adata2)
    sc.pp.normalize_total(adata3, target_sum=1e4)   
    sc.pp.log1p(adata3)
    sc.pp.normalize_total(adata4, target_sum=1e4)
    sc.pp.log1p(adata4)
    
    # Shuffle cells in adata1
    adata1 = adata1[np.random.permutation(adata1.n_obs), :]
    _, _, adata_res1= run_pipeline(adata1, epochs=1000, batch_size=2048, top_n_genes=5000, device='cuda:1')
    adata1 = adata1[adata1.obs['pseudo_angle'].sort_values().index, :]

    # Evaluate correlation score
    corr_score_perfect1 = calculate_correlation_score(adata1.obs.stage)
    print(corr_score_perfect1)
    
    # Construct periodic matrix and adjust PCA angle
    periodic_adata = get_periodic_matrix(adata1.copy(), top_k=10, use_highly_variable=False)
    adjust_adata1 = calculate_pca_angle(
        periodic_adata,
        use_rep='X',
        pca_key_added='X_pca_adjust',
        angle_key_added='pca_adjust_angle',
        plot=False
        )
    adjust_adata1 = adjust_adata1[adjust_adata1.obs.sort_values(by='pca_adjust_angle').index, :]
    corr_adjust_score_perfect1 = calculate_correlation_score(adjust_adata1.obs.stage)
    print(corr_adjust_score_perfect1)

    # Build reference dataset dictionary
    reference_dataset = {'ref_dataset1': adata2,
                        'ref_dataset2': adata3,
                        'ref_dataset3': adata4
                        }

    reference_res =  create_reference_gene_sets(reference_dataset)
    adata_adjust = calculate_gene_oscillation_properties_v2(adata1)
    cos_adata = reconstruct_cosine_expression(
        adata_adjust,
        min_frequency=1,
        max_frequency=2,
        min_amplitude=0.01)
    
    # Score cell cycle phases using multiple references
    score_multiple_g1_cell_cycle(
        adata=cos_adata,
        s_genes_list=[reference_res[t]['S'] for t in list(reference_dataset.keys())],
        g2m_genes_list=[reference_res[t]['G2M'] for t in list(reference_dataset.keys())],
        g1_genes_list=[reference_res[t]['G1'] for t in list(reference_dataset.keys())],
        prefix='ref'
    )

    # 1. Standardize cell cycle scores
    cycle_features = [
        'G1_score_ref1', 'S_score_ref1', 'G2M_score_ref1',
        'G1_score_ref2', 'S_score_ref2', 'G2M_score_ref2',
        'G1_score_ref3', 'S_score_ref3', 'G2M_score_ref3',
    ]

    # Only standardize non-missing values
    valid_mask = cos_adata.obs[cycle_features].notnull().all(axis=1)
    scaler = StandardScaler()
    scaled_values = scaler.fit_transform(cos_adata.obs.loc[valid_mask, cycle_features])

    # Store standardized values with '_scaled' suffix
    for i, col in enumerate(cycle_features):
        cos_adata.obs.loc[valid_mask, f'{col}_scaled'] = scaled_values[:, i]

    # 2. Predict phases for each reference
    phases = ['G1', 'S', 'G2M']
    refs = ['ref1', 'ref2', 'ref3']

    for ref in refs:
        # Predicted phase is the column with the maximum score
        cos_adata.obs[f'predicted_phase_{ref}'] = cos_adata.obs[
            [f'{phase}_score_{ref}' for phase in phases]
        ].idxmax(axis=1).str.replace('_score_' + ref, '')

    # 3. Create consensus prediction
    pred_cols = [f'predicted_phase_{ref}' for ref in refs]
    cos_adata.obs['predicted_cell_cycle_consensus'] = cos_adata.obs[pred_cols].mode(axis=1)[0]

    # 4. Filter for consistent predictions across all references
    phase_cols = ['predicted_phase_ref1', 'predicted_phase_ref2', 'predicted_phase_ref3']
    consistent_cells_mask = cos_adata.obs[phase_cols].nunique(axis=1) == 1
    cos_adata.obs['passed_consistency_filter'] = consistent_cells_mask

    # Compute average standardized score for each phase
    phases = ['G1', 'S', 'G2M']
    for phase in phases:
        cols = [f'{phase}_score_ref{i}_scaled' for i in range(1,4)]
        cos_adata.obs[f'{phase}_score_final'] = cos_adata.obs[cols].mean(axis=1)

    # Predict final cell cycle phase
    cos_adata.obs['predicted_phase_final'] = cos_adata.obs[
        ['G1_score_final', 'S_score_final', 'G2M_score_final']
    ].idxmax(axis=1).str.replace('_score_final', '')

    # Scale data (optional if not done earlier)
    sc.pp.scale(cos_adata, max_value=10)

    # Compute PCA with 10 components
    sc.tl.pca(cos_adata, svd_solver='arpack', n_comps=10)

    # Plot PCA colored by stage and scores
    sc.pl.pca(cos_adata, color=['stage', 'G1_score_final', 'S_score_final', 'G2M_score_final'])

    # Corrected scores
    cos_adata.obs['G1_corrected'] = 2*cos_adata.obs['G1_score_final'] - cos_adata.obs['S_score_final'] - cos_adata.obs['G2M_score_final']
    cos_adata.obs['S_corrected'] = 2*cos_adata.obs['S_score_final'] - cos_adata.obs['G1_score_final'] - cos_adata.obs['G2M_score_final']
    cos_adata.obs['G2M_corrected'] = 2*cos_adata.obs['G2M_score_final'] - cos_adata.obs['G1_score_final'] - cos_adata.obs['S_score_final']
    sc.pl.pca(cos_adata, color=['stage', 'S_score_final', 'G1_score_final', 'G2M_score_final', 'G1_corrected', 'S_corrected', 'G2M_corrected'], title='Stage')
    
        # Prepare features for clustering using corrected scores
    features_for_clustering = cos_adata.obs[['G1_corrected', 'S_corrected', 'G2M_corrected']].copy()

    # Check for NaN values, since GMM cannot handle NaN
    if features_for_clustering.isnull().values.any():
        print("Warning: NaN values detected in clustering features. Filling with 0. Consider a better imputation strategy.")
        features_for_clustering.fillna(0, inplace=True)  # Alternatively, use mean or other imputation methods

    # Initialize and fit Gaussian Mixture Model (GMM)
    n_clusters = 10  # Number of clusters (can be tuned)
    gmm = GaussianMixture(n_components=n_clusters, random_state=42, covariance_type='full')
    # covariance_type options: 'full', 'tied', 'diag', 'spherical'

    # Fit the model and predict cluster labels
    cluster_labels = gmm.fit_predict(features_for_clustering)

    # Add clustering results back to AnnData object
    # Convert numeric labels to categorical for better handling in Scanpy
    cos_adata.obs['GMM_clusters'] = pd.Categorical(cluster_labels)

    print(cos_adata.obs['GMM_clusters'].value_counts())

    # Visualize clustering results (recommended)
    # Use PCA plot with cluster labels
    if 'X_pca' not in cos_adata.obsm:
        print("Computing PCA for visualization...")
        sc.tl.pca(cos_adata, svd_solver='arpack')  # Use arpack for reproducibility

    sc.pl.pca(
        cos_adata,
        color=['stage', 'GMM_clusters',
               'G1_corrected', 'S_corrected', 'G2M_corrected'],
        title='PCA with True Stage and GMM Clusters (from corrected scores)',
        ncols=3,  # Adjust number of columns for better layout
    )

    # Ensure GMM clustering and corrected scores exist
    assert 'GMM_clusters' in cos_adata.obs, "Run GMM clustering first"
    required_scores = ['G1_corrected', 'S_corrected', 'G2M_corrected']
    assert all(score in cos_adata.obs for score in required_scores), "Missing corrected score columns"

    # Assign anchor cells based on specificity
    cos_adata = assign_anchor_cells(
        cos_adata,
        specificity_threshold=0.6,
        max_clusters_per_phase=1  # Limit to max 1 cluster per phase
    )

    # Split into source (domain1) and target (domain2)
    domain1 = cos_adata[cos_adata.obs['phase_anchor'] != 'None'].copy()
    domain2 = cos_adata[cos_adata.obs['phase_anchor'] == 'None'].copy()
    domain1 = periodic_adata[domain1.obs.index, ].copy()
    domain2 = periodic_adata[domain2.obs.index, ].copy()

    domain1.obs = cos_adata[domain1.obs.index, ].obs.copy()
    domain2.obs = cos_adata[domain2.obs.index, ].obs.copy()

    # Identify highly variable genes
    sc.pp.highly_variable_genes(periodic_adata, flavor='seurat', n_top_genes=1000)

    # Print number of highly variable genes
    print(f"Number of highly variable genes: {periodic_adata.var['highly_variable'].sum()}")

    # Extract top highly variable genes
    hvg_genes = periodic_adata.var[periodic_adata.var['highly_variable']].index.tolist()
    print(f"Top 10 highly variable genes: {hvg_genes[:10]}")

    # Subset domain1 and domain2 to highly variable genes
    domain1 = domain1[:, hvg_genes]
    domain2 = domain2[:, hvg_genes]

        # --- 1. Data Preparation ---
    # Ensure .X is a dense numpy array for PyTorch
    if hasattr(domain1.X, "toarray"):
        source_X = domain1.X.toarray().astype(np.float32)
    else:
        source_X = domain1.X.astype(np.float32)

    if hasattr(domain2.X, "toarray"):
        target_X = domain2.X.toarray().astype(np.float32)
    else:
        target_X = domain2.X.astype(np.float32)

    # Label encoding for source domain
    label_encoder = LabelEncoder()
    source_y_str = domain1.obs['phase_anchor'].astype(str)
    source_y = label_encoder.fit_transform(source_y_str)
    num_classes = len(label_encoder.classes_)
    print(f"Number of classes: {num_classes}")
    print(f"Class mapping: {dict(zip(label_encoder.classes_, range(num_classes)))}")


    BATCH_SIZE = 5  # Can be tuned

    source_dataset = AnnDataset(source_X, source_y)
    target_dataset = AnnDataset(target_X)  # No labels for target

    source_loader = DataLoader(source_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
    target_loader = DataLoader(target_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)

    # --- 3. Loss Functions and Optimizer ---
    device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    input_dim = source_X.shape[1]
    model = DANN(input_dim=input_dim, num_classes=num_classes).to(device)

    optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
    criterion_label = nn.CrossEntropyLoss()
    criterion_domain = nn.BCEWithLogitsLoss()  # For binary domain classification

    # --- 4. Training Loop ---
    N_EPOCHS = 100  # Can be tuned
    lambda_domain_initial = 0.1  # Initial weight for domain loss
    adapt_grl_alpha = True  # Whether to adapt GRL alpha during training

    print_every = 10

    for epoch in range(N_EPOCHS):
        model.train()
        len_dataloader = min(len(source_loader), len(target_loader))
        iter_source = iter(source_loader)
        iter_target = iter(target_loader)

        total_label_loss = 0
        total_domain_loss = 0
        total_source_acc = 0

        # Adapt GRL alpha (schedule increases domain adaptation importance over time)
        if adapt_grl_alpha:
            p = float(epoch) / N_EPOCHS
            current_alpha = 2. / (1. + np.exp(-10 * p)) - 1
            model.grl.set_alpha(current_alpha)
            current_lambda_domain = 1.0
        else:
            current_lambda_domain = lambda_domain_initial

        for i in range(len_dataloader):
            try:
                source_data, source_labels = next(iter_source)
            except StopIteration:
                iter_source = iter(source_loader)
                source_data, source_labels = next(iter_source)

            try:
                target_data, _ = next(iter_target)
            except StopIteration:
                iter_target = iter(target_loader)
                target_data, _ = next(iter_target)

            source_data, source_labels = source_data.to(device), source_labels.to(device)
            target_data = target_data.to(device)

            optimizer.zero_grad()

            # Source domain forward pass
            label_preds_source, domain_preds_source_logits = model(source_data)
            loss_label = criterion_label(label_preds_source, source_labels)

            # Domain loss for source (label 0)
            domain_labels_source = torch.zeros(source_data.size(0), 1).to(device)
            loss_domain_source = criterion_domain(domain_preds_source_logits, domain_labels_source)

            # Target domain forward pass (only domain prediction used)
            _, domain_preds_target_logits = model(target_data)
            domain_labels_target = torch.ones(target_data.size(0), 1).to(device)
            loss_domain_target = criterion_domain(domain_preds_target_logits, domain_labels_target)

            # Total domain loss
            loss_domain = loss_domain_source + loss_domain_target

            # Total loss
            total_loss = loss_label + current_lambda_domain * loss_domain

            total_loss.backward()
            optimizer.step()

            total_label_loss += loss_label.item()
            total_domain_loss += loss_domain.item()

            # Source accuracy
            _, predicted_labels_source = torch.max(label_preds_source, 1)
            total_source_acc += (predicted_labels_source == source_labels).sum().item()

        avg_label_loss = total_label_loss / len_dataloader
        avg_domain_loss = total_domain_loss / len_dataloader
        avg_source_acc = total_source_acc / (len_dataloader * BATCH_SIZE)

        if (epoch + 1) % print_every == 0 or epoch == 0:
            print(f"Epoch [{epoch+1}/{N_EPOCHS}], Alpha: {model.grl.alpha:.4f}, "
                  f"Label Loss: {avg_label_loss:.4f}, Domain Loss: {avg_domain_loss:.4f}, "
                  f"Source Acc: {avg_source_acc:.4f}")
            # --- 5. Prediction on Target Domain ---
    model.eval()
    target_predictions_list = []
    with torch.no_grad():
        # Re-create target loader without shuffling for ordered predictions
        target_loader_eval = DataLoader(target_dataset, batch_size=BATCH_SIZE, shuffle=False)
        for target_data, _ in target_loader_eval:
            target_data = target_data.to(device)
            output_target = model(target_data, mode='target_predict')
            _, predicted_target = torch.max(output_target, 1)
            target_predictions_list.extend(predicted_target.cpu().numpy())

    # Convert numerical predictions back to original string labels
    predicted_phase_labels = label_encoder.inverse_transform(target_predictions_list)

    # Add predictions to domain2 AnnData object
    # Ensure the length matches. If drop_last=True in DataLoader, some cells might be missed.
    if len(predicted_phase_labels) == domain2.n_obs:
        domain2.obs['predicted_phase_dann'] = predicted_phase_labels
        print("\nPredictions added to domain2.obs['predicted_phase_dann']")
        print(domain2.obs['predicted_phase_dann'].value_counts())
    else:
        print(f"\nWarning: Number of predictions ({len(predicted_phase_labels)}) "
              f"does not match number of target cells ({domain2.n_obs}). "
              "This may happen if drop_last=True was used in target_loader_eval "
              "or if BATCH_SIZE does not divide target_X.shape[0] evenly. "
              "Predictions not added to AnnData.")

    model.eval()
    with torch.no_grad():
        all_X = np.vstack([source_X, target_X])

        # Process in batches if memory is limited
        all_features_list = []
        temp_dataset = AnnDataset(all_X)
        temp_loader = DataLoader(temp_dataset, batch_size=BATCH_SIZE*2, shuffle=False)
        for batch_data, _ in temp_loader:
            batch_data = batch_data.to(device)
            features = model(batch_data, mode='features_only')  # Feature extraction mode
            all_features_list.append(features.cpu().numpy())

        all_features = np.concatenate(all_features_list, axis=0)

    # Create a temporary AnnData for visualization
    adata_for_viz = ad.AnnData(all_features)
    adata_for_viz.obs['domain'] = ['source'] * len(source_X) + ['target'] * len(target_X)
    source_labels_str = domain1.obs['phase_anchor'].tolist()

    # Handle cases where predicted labels are shorter than domain2.n_obs
    target_pred_labels_full = ['Unknown'] * domain2.n_obs
    target_pred_labels_full[:len(predicted_phase_labels)] = predicted_phase_labels
    adata_for_viz.obs['phase'] = source_labels_str + target_pred_labels_full
    adata_for_viz.obs['phase_numeric'] = label_encoder.transform(source_labels_str).tolist() + target_predictions_list[:len(target_pred_labels_full)]

    # Align labels for evaluation
    domain1.obs[['stage', 'phase_anchor']]
    domain2.obs[['stage', 'predicted_phase_dann']]
    domain2.obs.rename(columns={'predicted_phase_dann': 'test'}, inplace=True)
    domain1.obs.rename(columns={'phase_anchor': 'test'}, inplace=True)

    # Replace numeric stage values with string labels
    domain1.obs['stage'] = domain1.obs['stage'].replace({0: 'G1', 1: 'S', 2: 'G2M'})
    domain2.obs['stage'] = domain2.obs['stage'].replace({0: 'G1', 1: 'S', 2: 'G2M'})

    # Generate classification report
    true_labels = pd.concat([domain1.obs['stage'], domain2.obs['stage']])
    predicted_labels = pd.concat([domain1.obs['test'], domain2.obs['test']])

    report_dict = classification_report(true_labels, predicted_labels, output_dict=False)
    return report_dict, domain1, domain2


if __name__ == "__main__":
    # Example seeds
    res4, dataset1, dataset2 = run_analysis(SEED=79710)
```
